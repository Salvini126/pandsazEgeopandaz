```python
import contextily
```

```python
Duomo=Point(9.190589,45.463987)
```

```python
ax=milano[milano.contains(Duomo)].to_crs(epsg=3857).plot(figsize=(15,15),facecolor="none",edgecolor="k")
geopandas.GeoSeries([Duomo], crs="EPSG:4326").to_crs(epsg=3857).plot(ax=ax,color="red")
contextily.add_basemap(ax)
```

```python
milano[milano.touches(milano[milano.contains(Duomo)].geometry.squeeze())]
```

```python
vicinoDuomo=milano[milano.touches(milano[milano.contains(Duomo)].geometry.squeeze())]
ax=vicinoDuomo.to_crs(epsg=3857).plot(figsize=(15,15),facecolor="none",edgecolor="k")
contextily.add_basemap(ax)
```
