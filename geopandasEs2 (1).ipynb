```python
%matplotlib inline
import geopandas as gpd
import pandas as pd 
import matplotlib.pyplot as plt
import contextily
```

```python
content/drive/MyDrive/Limiti01012021_g/Reg01012021_g
```

```python
italia = gpd.read_file("/content/drive/MyDrive/Limiti01012021_g/Reg01012021_g")
italia
```

```python
ax = italia.to_crs(epsg=3857).plot(figsize=(12,10), facecolor= "none", edgecolor = "b")
contextily.add_basemap(ax)
```

```python
ax = italia.to_crs(epsg=3857).plot(figsize=(12,10), facecolor= "none", edgecolor = "b")
contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
```

```python
province = gpd.read_file("/content/drive/MyDrive/Limiti01012021_g/ProvCM01012021_g")
province
```

```python
lombardia = italia.loc[italia['DEN_REG'] == 'Lombardia', 'geometry'].squeeze()
lombardia
```

```python
proLomb = province[province.within(lombardia)]
proLomb
```

```python
ax = proLomb.to_crs(epsg=3857).plot(figsize=(20,10), facecolor= "none", edgecolor= "k")
contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
```

```python
comuni = gpd.read_file("/content/drive/MyDrive/Limiti01012021_g/Com01012021_g")
comuni
```

```python
milano = province.loc[province['DEN_CM'] == 'Milano', 'geometry'].squeeze()
milano
```

```python
comuniDiMilano = comuni[comuni.within(milano)]
ax = comuniDiMilano.to_crs(epsg= 3857).plot(figsize= (12,15), facecolor = "none", edgecolor = "b")
contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
```

```python
quartieri = gpd.read_file("/content/drive/MyDrive/ds964_nil_wm.zip")
quartieri
```
